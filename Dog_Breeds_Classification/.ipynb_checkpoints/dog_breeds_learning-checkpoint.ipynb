```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf

# CUDA를 사용할 수 있는지 확인
print("Is CUDA available?", tf.test.is_built_with_cuda())

# 사용 가능한 GPU 디바이스 출력
print("Available GPUs:", tf.config.list_physical_devices('GPU'))

# 현재 사용 중인 디바이스 출력
print("Current device:", tf.test.gpu_device_name())

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
#breed_list = os.listdir('./test')[:10]
breed_list = ['Shih-Tzu', 'Maltese', 'Yorkie', 'Poodle', 'Golden Retriever', 'Bulldog', 'Corgi', 'Beagle']
breed_list
```

```python
img_count = 0
for breed in breed_list:
    img_count += len(os.listdir('./train/'+breed))
img_count
train_image = np.zeros((img_count, 224, 224, 3))
train_label = np.zeros((img_count, len(breed_list)))
i = 0
j = 0
for breed in breed_list:
    for img_name in os.listdir('./train/'+breed):
        train_image[i] = np.array(Image.open('./train/'+breed+'/'+img_name))
        train_label[i][j] = 1
        i += 1
    j += 1
plt.imshow(train_image[0]/255)
train_label[0]
```

```python
img_count = 0
for breed in breed_list:
    img_count += len(os.listdir('./valid/'+breed))
print(img_count)

valid_image = np.zeros((img_count, 224, 224, 3))
valid_label = np.zeros((img_count, len(breed_list)))
i = 0
j = 0
for breed in breed_list:
    for img_name in os.listdir('./valid/'+breed):
        valid_image[i] = np.array(Image.open('./valid/'+breed+'/'+img_name))
        valid_label[i][j] = 1
        i += 1
    j += 1
plt.imshow(valid_image[0]/255)
valid_label[0]
```

```python
img_count = 0
for breed in breed_list:
    img_count += len(os.listdir('./test/'+breed))
print(img_count)

test_image = np.zeros((img_count, 224, 224, 3))
test_label = np.zeros((img_count, len(breed_list)))
i = 0
j = 0
for breed in breed_list:
    for img_name in os.listdir('./test/'+breed):
        test_image[i] = np.array(Image.open('./test/'+breed+'/'+img_name))
        test_label[i][j] = 1
        i += 1
    j += 1
plt.imshow(test_image[0]/255)
test_label[0]
```

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.2),
])
```

```python
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')
```

```python
feature_batch = base_model(valid_image[:10])
print(feature_batch.shape)
```

```python
base_model.trainable = False
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
```

```python
prediction_layer = tf.keras.layers.Dense(len(breed_list))
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
```

```python
inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
```

```python
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.summary()
```

```python
len(model.trainable_variables)
```

```python
history = model.fit(train_image,
                    train_label,
                    batch_size=20,
                    epochs=15,
                    validation_data=(valid_image, valid_label))
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
base_model.trainable = True
```

```python
# Let's take a look to see how many layers are in the base model
print("Number of layers in the base model: ", len(base_model.layers))

# Fine-tune from this layer onwards
fine_tune_at = 100

# Freeze all the layers before the `fine_tune_at` layer
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False
```

```python
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              optimizer = tf.keras.optimizers.RMSprop(learning_rate=base_learning_rate/10),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(model.trainable_variables)
```

```python
fine_tune_epochs = 10
total_epochs =  15 + fine_tune_epochs

history = model.fit(train_image,
                    train_label,
                    batch_size=20,
                    epochs=total_epochs,
                    initial_epoch=history.epoch[-1],
                    validation_data=(valid_image, valid_label))
```

```python
acc += history.history['accuracy']
val_acc += history.history['val_accuracy']

loss += history.history['loss']
val_loss += history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.ylim([0.8, 1])
plt.plot([15-1,15-1],
          plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.ylim([0, 1.0])
plt.plot([15-1,15-1],
         plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
loss, acc = model.evaluate(test_image, test_label)
print("Untrained model, accuracy: {:5.2f}%".format(100*acc))
```
